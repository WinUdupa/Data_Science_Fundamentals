## Making figures with Numpy

```python
import numpy as np
```

```python
x = np.linspace(0, 100, 10, dtype = int)
x
```

```python
y = np.linspace(0, 500, 10, dtype = float)
y
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
ax.plot(x, y)
ax.set(title = "Numpy",
       xlabel = "x axis",
       ylabel = "y axis")

fig.savefig("image/numpy_sample.png")
plt.show()
```

```python
# Using same data plotting scatter plot and hist and bar
fig, ax = plt.subplots()
ax.scatter(x, y, color = 'red')

ax.set(title = "Numpy Scatter",
       xlabel = "x axis",
       ylabel = "y axis")

fig.savefig("image/numpy_scatter.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'green')

ax.set(title = "Numpy bar",
       xlabel = "x axis",
       ylabel = "y axis")

fig.savefig("image/numpy_bar.png") #Bar plot will not take x,y instead it takes x and height (y is the height here)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(x, y, color = 'green')

ax.set(title = "Numpy hist",
       xlabel = "x axis",
       ylabel = "y axis")

fig.savefig("image/numpy_hist.png")
plt.show()
```

## Options for Subplot

```python
# Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                              figsize = (10, 5))

ax1.plot(x, y)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3. bar(np.random.random(10), np.random.random(10))
ax4.hist(x, y)
plt.show()
```

```python
# Option 2

fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

ax[0, 0].plot(x, y)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0]. bar(np.random.random(10), np.random.random(10))
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

## Plotting From Pandas DataFrame

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex = True)
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Date"] = pd.date_range("1/1/2025", periods = 10) #or can also be len(ccar_sales)
car_sales
```

```python
# Converting the Price column to int
car_sales["Price"] = car_sales["Price"].astype(int)
del car_sales["Prices"]
```

```python
car_sales["Price"][0] + car_sales["Price"][1]
```

```python
car_sales
```

```python
Total = car_sales["Price"].cumsum()
Total
```

```python
car_sales["total"] = Total
car_sales
```

```python
#Plotting Total sales 
fig, ax = plt.subplots(figsize = (10, 5))
x = car_sales["Date"]
y = car_sales["total"]
ax.plot(x, y)
plt.show()
```

```python
car_sales.plot(x = "Date", y = "total")
plt.show()
```

```python
car_sales.plot(x = "Odometer (KM)", y = "Doors", kind = "bar")
plt.show()
```
